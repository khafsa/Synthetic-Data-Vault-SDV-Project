```python
import pandas as pd

path = "/content/drive/MyDrive/Colab Notebooks/data/Responses_31.csv"
data = pd.read_csv(path)

# Dataset is now stored in a Pandas Dataframe
```

```python
data.head()
```

```python
from sdv.tabular import CTGAN
```

```python
model = CTGAN()
```

```python
model.fit(data)
```

```python
new_data = model.sample(num_rows=40)
```

```python
new_data
```

```python
model.save('my_model1.csv')
```

```python
loaded = CTGAN.load('my_model1.csv')
new_data = loaded.sample(num_rows=40)
```

Specifying the Primary Key of the table


```python
data.Case_No.value_counts().max()
```

```python
new_data[new_data.Case_No == new_data.Case_No.value_counts().index[0]]
```

```python
model = CTGAN(
    primary_key='Case_No'
)
model.fit(data)
new_data = model.sample(40)
new_data.head()
```

```python
new_data
```

```python
from google.colab import files

new_data.to_csv('df2.csv')
files.download('df2.csv')
```

Save and share the model

```python
model.save('my_model2.csv')
```

Load the model and generate new data

```python
loaded = CTGAN.load('my_model2.csv')
new_data1 = loaded.sample(num_rows=200)
```

Specifying the Primary Key of the table

```python
data1.Case_No.value_counts().max()
```

```python
new_data1[new_data1.Case_No == new_data1.Case_No.value_counts().index[0]]
```

As a result, the model will learn that this column must be unique and generate a unique sequence of values for the column:


```python
model = CTGAN(
    primary_key='Case_No'
)
model.fit(data1)
new_data1 = model.sample(200)
new_data1.head()
```
